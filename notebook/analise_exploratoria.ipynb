```python
import pandas as pd
import os

#Configuração para exibir todas as colunas no terminal

pd.set_option('display.max_columns', None)
pd.set_option('display.width', 1000)

def carregar_dados():
    # Caminho base (ajuste se necessário)
    base_path = "../data/raw"
    
    print("--- Carregando Datasets ---")
    
    # 1. Tabela de Pedidos (Datas e Status)
    try:
        print("Carregando Orders...")
        df_orders = pd.read_csv(os.path.join(base_path, 'olist_orders_dataset.csv'))
        print(f"Orders carregado: {df_orders.shape[0]} linhas")
    except FileNotFoundError:
        print("ERRO: Arquivo 'olist_orders_dataset.csv' não encontrado.")
        return None, None, None

    # 2. Tabela de Itens (Preço e Produto)
    try:
        df_items = pd.read_csv(os.path.join(base_path, 'olist_order_items_dataset.csv'))
        print(f"Items carregado: {df_items.shape[0]} linhas")
    except FileNotFoundError:
        print("ERRO: Arquivo 'olist_order_items_dataset.csv' não encontrado.")

    # 3. Tabela de Clientes (IDs e Localização)
    try:
        df_customers = pd.read_csv(os.path.join(base_path, 'olist_customers_dataset.csv'))
        print(f"Customers carregado: {df_customers.shape[0]} linhas")
    except FileNotFoundError:
        print("ERRO: Arquivo 'olist_customers_dataset.csv' não encontrado.")

    return df_orders, df_items, df_customers

# Execução Principal
if __name__ == "__main__":
    orders, items, customers = carregar_dados()
    
    if orders is not None:
        print("\n--- Amostra: Pedidos (Orders) ---")
        print(orders.head(3))
        print("\n--- Colunas Importantes em Orders ---")
        print(orders.columns.tolist())
        
        print("\n--- Amostra: Itens (Items) ---")
        print(items.head(3))
        
        print("\n--- Amostra: Clientes (Customers) ---")
        print(customers.head(3))
```
